## Image Segment Activations

Here we examine how specific image/facial segments contribute to age estimations, and how this interacts with racial categorisations.

```python
import json
import numpy as np
import pandas as pd
from PIL import Image
import seaborn as sns
import matplotlib.pyplot as plt
plt.ioff()
%matplotlib inline
from analysis_3_filter_visualise import *

image_masks_fpath = "utkface/img-masks/"
utkface_data = json.loads(open("metadataset_utkface_activations.json").read())
utkface_data_face_vs_outer, utkface_data_contribution_pct = generate_processed_datasets(utkface_data)

def visualise_heatmaps(this_type="accurate", this_age_range=[0, 12]):
    heatmaps = visualise_heatmap_corr_pairplot_all(filter_dataset(utkface_data_face_vs_outer["positive"], 
        KEYS_RACE_AGE_BOUNDARIES, "race_real", RACES, age_min=this_age_range[0], age_max=this_age_range[1]), 
        image_masks_fpath, x_vars=KEYS_BOUNDARIES, pairplot_scatter_alpha=0.025, 
        dataset_tuple2=filter_dataset(utkface_data_face_vs_outer["negative"], 
        KEYS_RACE_AGE_BOUNDARIES, "race_real", RACES, 
        age_min=this_age_range[0], age_max=this_age_range[1]),this_type=this_type)
```
